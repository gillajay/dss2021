# SIMULATED OBSERVATION 

Note: Flat fielding is not included in this simulation

```python
readnoise = 6 #e-
background = 10 #e- (sky background)
darkcurrent = 1 #e-
biasoffset = 1000 #e- (offset value set by electronics)
imagesize = 201

nbiasframes = 20
ndarkframes = 20

bias = np.random.normal(size=(imagesize,imagesize),loc=biasoffset,scale=readnoise)
bkg = np.random.poisson(size=(imagesize,imagesize),lam=background)
dark = np.random.poisson(size=(imagesize,imagesize),lam=darkcurrent)

allbias = np.random.normal(size=(imagesize,imagesize,nbiasframes),loc=biasoffset,scale=readnoise)
masterbias = np.median(allbias,axis=2)

alldark = np.random.poisson(size=(imagesize,imagesize,ndarkframes),lam=darkcurrent)
masterdark = np.median(alldark,axis=2)
```

```python
def gaussianimg(size, x0, y0, intensity, sigma, nonoise=False):
    img = np.zeros((size,size))
    totflux = 0
    for i in range(size):
        for j in range(size):
            val = intensity*math.exp(-4*math.log(2) * ((i-x0)**2 + (j-y0)**2) / sigma**2)
            totflux += val
            if (nonoise):
                img[i,j] = val
            else:    
                img[i,j] = np.random.poisson(lam=val)
    print('Total True Flux: %f'%totflux)        
    return(img)
```

```python
brightness = 50
width = 3
starnonoise = gaussianimg(imagesize,100,100,brightness,width,nonoise=True)
plt.imshow(starnonoise[75:125,75:125],interpolation='none')
plt.colorbar();
```

```python
star = gaussianimg(201,100,100,brightness,width)
myobs = star + bias + bkg + dark - masterbias - masterdark
```

```python
plt.imshow(star[75:125,75:125],interpolation='none')
plt.colorbar();
```

```python
plt.imshow(myobs[75:125,75:125],interpolation='none')
plt.colorbar();
```

```python
def aperturephot(img,x,y,dr):
    dbinner = 15
    dbouter = 20
    drmax = 25
 
    appix = 0.0
    nappix = 0.0
    bkgpix = 0.0
    nbkgpix = 0.0
    
    bkgpixval = []
    
    for i in np.arange(x-drmax,x+drmax):
        for j in np.arange(y-drmax,y+drmax):
            if((i-x)**2+(j-y)**2<=dr**2):
                appix += img[i,j]
                nappix += 1 
            if(((i-x)**2+(j-y)**2<=dbouter**2) and ((i-x)**2+(j-y)**2>dbinner**2)):
                bkgpix += img[i,j]
                bkgpixval.append(img[i,j])
                nbkgpix += 1
    
    flux = appix - bkgpix*(nappix/nbkgpix)
    
    return(flux,bkgpix/nbkgpix,nappix,nbkgpix,np.var(np.array(bkgpixval),ddof=1))
```

```python
fluxnonoisevals = np.zeros(15)
fluxvals = np.zeros(15)
bkgvals = np.zeros(15)
nappix = np.zeros(15)
nbkgpix = np.zeros(15)
bkgvarvals = np.zeros(15)

for i in range(len(fluxvals)):
    fluxnonoisevals[i], _, _, _, _ = aperturephot(starnonoise,100,100,i)
    fluxvals[i],bkgvals[i],nappix[i],nbkgpix[i],bkgvarvals[i] = aperturephot(myobs,100,100,i)
```

```python
plt.plot(fluxnonoisevals,'o-')
plt.plot(fluxvals,'o-')
plt.xlabel('Aperture radius (pix)')
plt.ylabel('Flux')
print('Total Flux: %f'%fluxvals[12])
```

# SNR of Aperture Photometry

```python
noise = np.sqrt(fluxvals + nappix*(bkgvals + readnoise**2)+(nappix/(nbkgpix*1.0))*(bkgvals + readnoise**2))
plt.plot(noise,'o-')
```

```python
snra = fluxvals/np.sqrt(fluxvals + nappix*(bkgvals + readnoise**2)+(nappix/(nbkgpix*1.0))*(bkgvals + readnoise**2))
plt.plot(snra,'o-')
plt.xlabel('Aperture radius (pix)')
plt.ylabel('SNR');
plt.savefig('snr.png',dpi=300)
```

```python
correction = fluxvals[2]/fluxvals[14]
print('Aperture Correction: %f'%correction)
```

```python
snrb = fluxvals/np.sqrt(fluxvals+nappix*bkgvarvals+(nappix/(nbkgpix*1.0)*bkgvarvals))
plt.plot(snra,'o-')
plt.plot(snrb,'o-');
```

```python
estflux = fluxvals[2]/correction
print('Estimated Flux: %f' % estflux)
```
